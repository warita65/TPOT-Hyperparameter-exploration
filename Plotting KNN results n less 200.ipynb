### Plotting k chosen by tpot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pmlb import fetch_data
import seaborn as sns
import scipy.stats
%matplotlib inline
```

# Seed 1

```python
data = pd.read_csv('Downloads/seed1.csv')
df = data.sort_values(by=['n_neighbors'])
```

### Plot of n_observations and n_neighbors

```python
X = df['n_observations']
y = df['n_neighbors']
```

```python
sns.regplot(X,y, 
            data=df);
plt.savefig('seed1_plot_observations.png')
```

#### Correlation value

```python
#correlation between the number of observations and the number of n_neighbors 
#Pearson's linear correlation

r_observations1 = np.corrcoef(X, y)
r_observations1
```

### Plot of n_features and n_neighbors

```python
X_f = df['n_features']
y = df['n_neighbors']
```

```python
sns.regplot(X_f,y, 
            data=df);
plt.savefig('seed1_plot_features.png')
```

#### Correlations

```python
r_observations1 = np.corrcoef(X_f, y)
r_observations1
```
