# TPOT using KNN 

```python
import pandas as pd
import numpy as np
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
from pmlb import classification_dataset_names
from pmlb import fetch_data
```

### Banana

```python
banana = fetch_data('banana', local_cache_dir = './')
```

```python
n_observations = banana.size
```

```python
features_banana = banana.drop('target', axis = 1)
```

```python
train_banana_features, test_banana_features, train_banana_target, test_banana_target = train_test_split(features_banana, banana['target'], train_size= 0.75, test_size = 0.25, random_state=42)
```

```python
#KNN with default values
tpot_config = {
  'sklearn.neighbors.KNeighborsClassifier': {
        'n_neighbors': range(1,101),
        'weights': ["uniform", "distance"],
        'p': [1, 2]
    }
}
```

```python
tpot = TPOTClassifier(generations=100, population_size=100, verbosity=2,
                       random_state = 42, config_dict=tpot_config, template = 'Classifier')
tpot.fit(train_banana_features, train_banana_target)
print(tpot.score(test_banana_features, test_banana_target))
tpot.export('tpot_KNN_banana3.py')
```

```python
#KNN with default values
tpot_config = {
  'sklearn.neighbors.KNeighborsClassifier': {
        'n_neighbors': range(1,int(n_observations/2)),
        'weights': ["uniform", "distance"],
        'p': [1, 2]
    }
}
```

```python
tpot = TPOTClassifier(generations=100, population_size=100, verbosity=2,
                       random_state = 42, config_dict=tpot_config , template = 'Classifier')
tpot.fit(train_banana_features, train_banana_target)
print(tpot.score(test_banana_features, test_banana_target))
tpot.export('tpot_KNN_banana_half2.py')
```
