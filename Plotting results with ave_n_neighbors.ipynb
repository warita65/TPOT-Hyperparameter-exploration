# Plot of average n_neighbors 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pmlb import fetch_data
import seaborn as sns
import scipy.stats
%matplotlib inline
```

```python
data = pd.read_excel('Downloads/ave_neighbors.xlsx')
data.head()
data.isnull()
df = data.sort_values(by=['n_features'])
```

```python
df.dropna()
```

```python
df.head(100)
```

```python
df = df.dropna(axis=0, subset=['ave_n_neighbors'])
```

```python
df.head(100)
```

### Plot n_observations and ave_n_neoghbors

```python
X = df['n_observations']
y = df['ave_n_neighbors']
```

```python
sns.regplot(X,y, 
            data=df);
plt.savefig('plot_ave&observation.png')
```

### Correlation between n_observations and ave_n_neighbor

```python
#correlation between the number of observations and the number of n_neighbors 
#Pearson's linear correlation

r_observations = np.corrcoef(X,y)
r_observations
```
